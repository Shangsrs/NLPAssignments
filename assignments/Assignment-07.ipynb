#### 切词

```python
content_id = []
content_source = []
content_source_ = []
content_text = []
for i, content in enumerate(news['content']):    
    id, source, news_content = news['id'][i], news['source'][i], news['content'][i]
    text = str(news_content)
    text = text.strip()
    if not text: continue
    text = langconv.Converter('zh-hans').convert(text)
    text = ''.join(re.findall(r'[^\\n]', text))
    text = ''.join(re.findall(r'[\d|\w]+',text))
    text = ' '.join(jieba.cut(text))
    content_id.append(id)    
    content_text.append(text)
    content_source.append(source)
    if i % 2000 ==0:
        print('Saved ' + str(i) + ' lines')    
```

#### 结果评估

```python
from sklearn import metrics 
```

```python
def measures(pred):    
    print("accuracy      : %0.3f" % metrics.accuracy_score(test_y, pred))
    print("precision score: %0.3f" % metrics.precision_score(test_y, pred))
    print("recall score  : %0.3f" % metrics.recall_score(test_y, pred))
    print("f1 score      : %0.3f" % metrics.f1_score(test_y, pred))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_pred = lr.fit(train_x,train_y).predict(test_x)
measures(lr_pred)
```

### SVM

```python
from sklearn.svm import LinearSVC
```

```python
svm = LinearSVC()
svm_pred = svm.fit(train_x, train_y).predict(test_x)
measures(svm_pred)
```
