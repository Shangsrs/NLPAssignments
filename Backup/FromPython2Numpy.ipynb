```python
import random
import matplotlib.pyplot as plt
```

```python
class RandomWalker:
    def __init__(self):
        self.position = 0

    def walk(self, n):
        self.position = 0
        for i in range(n):
            yield self.position
            self.position += 2*random.randint(0, 1) - 1

walker = RandomWalker()
```

```python
walk = [position for position in walker.walk(1000)]
x = list(range(len(walk)))
plt.plot(x,walk)
```
